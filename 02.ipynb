```python
import requests
from bs4 import BeautifulSoup
google_url = 'https://www.google.com.tw/search'
my_params = {'q': 'mlb'}
r = requests.get(google_url, params = my_params)
if r.status_code == requests.codes.ok:
    soup = BeautifulSoup(r.text, 'html.parser')
    print(soup.prettify())
  
```
